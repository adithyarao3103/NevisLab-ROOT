# Analysing Tree

```python
import ROOT
```

```python
ebeamhist = ROOT.TH1D("ebeamHist", "Histogram for ebeam", 100, 149.2, 150.8)
zvhist = ROOT.TH1D("zvHist", "Histogram for zv {zv < 20}", 100, 0, 30)
c = ROOT.TCanvas()
```

```python
myfile = ROOT.TFile('../DATA/experiment.root')
mychain = ROOT.gDirectory.Get('tree1')
entries = mychain.GetEntriesFast()
```

```python
for jentry in range(entries):
    nb = mychain.GetEntry(jentry)
    if nb <= 0:
        continue
    ebeam = mychain.ebeam
    ebeamhist.Fill(ebeam)
    zv = mychain.zv
    if (zv < 20):
        zvhist.Fill(zv)    

```

```python
ebeamhist.Draw()
c.Draw()
```

```python
zvhist.Draw()
c.Draw()
```
