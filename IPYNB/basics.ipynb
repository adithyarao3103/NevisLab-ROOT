```python
import ROOT
```

```python
c = ROOT.TCanvas()
```

## Plotting a Function

```python
f1 = ROOT.TF1("func1", "sin(x)/x", 0, 10)
f1.Draw()
c.Draw()
```

## Excercise 1 : Plotting a 2D function

```python
f2 = ROOT.TF2("func2", "sin(x)*sin(y)/(x*y)", -10, 10, -10, 10)
f2.Draw("surf1")
c.Draw()
```

## Histogram

```python
h1 = ROOT.TH1D("hist1", "Example Histogram", 100, -3, 3)
h1.FillRandom("gaus", 10000)
h1.Draw()
c.Draw()
```

## Custom Histogram

```python
myfun = ROOT.TF1("myfun", "sin(x)", 0, ROOT.TMath.Pi())
h = ROOT.TH1D("hist1", "Custom Histogram", 100, 0, ROOT.TMath.Pi())
h.FillRandom("myfun", 10000)
h.Draw("e")
c.Draw()
```

## Fitting Histogram to Gaus

```python
c2 = ROOT.TCanvas()
```

```python
h_FIT = ROOT.TH1D("hist_fit", "Histogram", 100, -3, 3)
h_FIT.FillRandom("gaus", 10000)
h_FIT.Fit("gaus")
h_FIT.Draw()
c2.Draw()
ROOT.gStyle.SetOptFit(0o0001)
```

```python
file1 = ROOT.TFile('../DATA/histogram.root')
h_cf = file1.Get('hist2')
fitFun = ROOT.TF1('fitFun','gaus(0) + gaus(3)', 0, 14)
fitFun.SetParameters(5., 2., 1., 1., 10., 1.)
h_cf.Fit("fitFun")
h_cf.Draw()
c2.Draw()
```
