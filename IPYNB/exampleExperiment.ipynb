```python
import ROOT
%jsroot
```

```python
experiment = ROOT.TFile.Open("../DATA/experiment.root")
tree1 = experiment.Get("tree1")
tree1.Print()
```

## Scatter Plot

This plots a scatter plot of the data. Correlations in data appear as structures in this plot

```python
c = ROOT.TCanvas()
```

```python
tree1.Draw("pz:zv")
c.Draw()
```

## Cuts
One can apply cuts on the data to select only the data that is of interest.

```python
tree1.Draw("zv", "zv<20")
c.Draw()
```

## Cuts applied to other variables

```python
tree1.Draw("ebeam", "zv<20")
c.Draw()
```

## Merging two cuts

```python
tree1.Draw("ebeam", "zv<20 && px>10")
c.Draw()
```
